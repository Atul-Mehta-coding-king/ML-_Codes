```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0,10.0)

#reading Data

data = pd.read_csv('//content/headbrain.csv')
```

```python
print(data.head(10))
data.shape
```

```python
# Let's collecting the data
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values

```

```python
# Now we need to find the values of b1 and b0 for equation y = b1x + b0
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values
m = len(x)

# let's excute the values of b1 and b0

numer = 0 
denom = 0

for i in range(m):
  numer  += (x[i]-mean_x) * (y[i]-mean_y)
  denom += (x[i]-mean_x)**2
b1 = numer / denom
b0 = mean_y - (b1*mean_x)

#print cofficient

print(b1 , b0)

```

```python
# plotting values and regression line

max_x = np.max(x) + 100
min_x = np.min(x) - 100

# calculating line values x and y
X = np.linspace(min_x , max_x , 1000)
Y = b0 + b1 * X

# plotting line

plt.plot(X , Y,color = "#58b970" , label = "Regression_line")
#plotting scatter plot
plt.scatter(x,y , c = "#ef5432" , label = "Scatter plot")

plt.xlabel('Head size of Brain in cm^3')
plt.ylabel('Brain weight in grams')
plt.legend()
plt.show()
```

```python
# let's calculate R^2 values to improve our model

ss_t = 0
ss_r = 0

for i in range(m):
  Y_pred = b0 + b1 * x[i]
  ss_t += (y[i]-mean_y)**2
  ss_r += (y[i]-Y_pred)**2

r2 = 1 - (ss_r/ss_t)
r2
```
